```python
#4.Fill in all the missing values present in all the columns in the dataset
data.fillna(method='ffill', inplace=True)
```

```python
data.head()
```

```python
#5.Check and handle outliers in at least 3 columns in the dataset
# Handling outliers in 'Age' column
age_mean = data['Age'].mean()
age_std = data['Age'].std()
age_cutoff = age_mean + 3 * age_std
data.loc[data['Age'] > age_cutoff, 'Age'] = age_cutoff
```

```python
# Handling outliers in 'Fare' column
fare_cutoff = data['Fare'].quantile(0.90)
data.loc[data['Fare'] > fare_cutoff, 'Fare'] = fare_cutoff
```

```python

# Handling outliers in 'SibSp' column
sibsp_cutoff = 4
data.loc[data['SibSp'] > sibsp_cutoff, 'SibSp'] = sibsp_cutoff
```
